## EP:
- 通过math库，写一个程序，使得用户输入三个顶点(x,y)返回三个角度
- 注意：Python计算角度为弧度制，需要将其转换为角度
<img src="../Photo/18.png">

```python
import math 
x1,y1= eval(input('输入顶点以,分隔'))
x2,y2= eval(input('输入顶点以,分隔'))
x3,y3= eval(input('输入顶点以,分隔'))
a=math.sqrt((x3-x2)**2+(y2-y3)**2)
c=math.sqrt((x2-x1)**2+(y2-y1)**2)
b=math.sqrt((x3-x1)**2+(y3-y1)**2)
A=math.degrees(math.acos((a*a-b*b-c*c) / (-2*b*c)))
B=math.degrees(math.acos((b*b-a*a-c*c) / (-2*a*c)))
C=math.degrees(math.acos((c*c-a*a-b*b) / (-2*a*b)))
print('A的角度为' + str(round(A))+'度   B的角度为'+str(round(B))+'度   C的角度为'+str(round(C))+'度')
```

## 转义序列 \
- a = "He said,"Johon's program is easy to read"
- 转掉它原来的意思
- 一般情况下只有当语句与默认方法相撞的时候，就需要转义

```python
long = "He said,\"Johon's program is easy to read"
for i in long :
    print(chr(ord (i)+10),end='')     #加密
```

```python
long = "He said,\"Johon's program is easy to read"
res=''
for i in long :
    res +=chr(ord (i)+10)
print(res)                            #加密
```

```python
import hashlib
str_ = 'wjg123' 
h1 = hashlib.md5()    #创建md5对象
h1.update(str_.encode(encoding='utf-8'))
print('MD5加密前为：'+str_)
print('MD5加密后为：'+h1.hexdigest())  
```

```python
import hashlib    #导入hashlib函数
dataset = '97c2f3947dd25ef98151e9b0dbd32a3a'   #数据库中的密码
password=input('请输入密码')  
hx = hashlib.md5()
hx.update(password.encode(encoding='utf-8'))  #输入的密码进行MD5加密，更新
if dataset==hx.hexdigest():   #数据库中的密码和加密后对比
    print('登录成功')
else:
    print('登录失败')
```
